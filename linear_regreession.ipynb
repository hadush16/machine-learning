```python
!git clone https://github.com/hadushb/machine-learning.git
import matplotlib.pyplot as plt
from scipy import stats

x = [5, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6]
y = [99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85] # Removed the extra element to match the length of x

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
    return slope * x + intercept

mymodel = list(map(myfunc, x))
plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
radio_spend = np.array([10, 15, 20, 25, 30, 35, 40, 45, 50, 55])  # Radio advertising spend in thousands of dollars
units_sold = np.array([150, 180, 210, 240, 270, 300, 330, 360, 390, 420])  # Units sold
```

```python
slope, intercept = np.polyfit(radio_spend, units_sold, 1)
# This function returns an array containing the polynomial coefficients,
# in this case, the slope and y-intercept of the line.

print("Slope (m):", slope)
print("Y-intercept (b):", intercept)
```

```python
# Create a scatter plot of the data points
plt.scatter(radio_spend, units_sold, label="Data Points")

# Create the linear regression line
regression_line = slope * radio_spend + intercept

# Plot the regression line
plt.plot(radio_spend, regression_line, color='red', label="Regression Line")

# Customize the plot
plt.xlabel("Radio Advertising Spend (in thousands of dollars)")
plt.ylabel("Units Sold")
plt.title("Linear Regression of Radio Advertising Spend vs Units Sold")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
